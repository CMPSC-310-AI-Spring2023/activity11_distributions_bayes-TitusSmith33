## Binomial Distribution

- The random.binomial function from numpy draws samples from a binomial distribution. The parameter n is an int of of the number of trials, p is a float that represents the probability of success. The last parameter, s, is an optional parameter for this function that handles the size that you wish for this function to work within.

- The plt are all formatting tools used for creating output in a specific manner.

- The binomial output shows a data point where a lot of the data sets towards the middle and you do not see a lot of waves abover or under 60/40. The data sits around 50%.

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

- Tge random.uniform function from numpy draws samples from a uniform distribution. The parameters low and high set the boundaries for this program to work within, and it is likely that every value between the two values is drawn. This function uses the low value, but excluded the high value. Again, the final parameter is an option parameter that sets the specified size of the run.

- The matplot functions allows you customize how your output is displayed.

- The data is a lot more scattered and wavy comapred to the binomial

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

- The numpy function random.normal draws samples from a Gaussian distribution. The mu parameter is the mean, the sigma parameter is the standard deviation, and again the final parameter is the optional size variable.

- The matplot functions allows you customize how your output is displayed.

- This data was unique because it all sits down towards the bottom without a lot of waves, but there are some large peaks.

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

## Bayes Algorithm

- Half of the data is set for testing, half for training (test_size = 0.5)
-Experiment this value

-Ideally you want around 75 for training and 25 for testing

Ref: https://scikit-learn.org/stable/modules/naive_bayes.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))

#gnb.predict_proba(X_test)
```
